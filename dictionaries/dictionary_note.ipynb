## Dictionaries 
dictionaries is a data structure that stores data as key - value paires

**A dictionary is a mapping**

```python
numbers = {}    #initializing an empty dictionary
numbers
```

```python
print(type(numbers))    #type class dict
```

### Adding items to the dictionary
this is done by using the bracket operator []


```python
numbers["one"] = 1    # the value in the [ ] is the key anf the value after the = is the value
numbers["two"] = 2
```

```python
print(numbers)    # Printing a dictionary
```
