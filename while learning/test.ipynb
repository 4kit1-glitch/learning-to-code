```python
from jupyturtle import make_turtle, forward, left, penup, pendown

make_turtle(width=300, height=300)

def line(lenght):
    forward(lenght)

def square(lenght):
    for i in range(4):
        line(lenght)
        left(90)
    else:
        print("a square drawn")

def circle():
    for i in range(30):
        line(5)
        left(45)

square(40)
square(90)
square(100)
circle()
```
